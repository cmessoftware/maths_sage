**Por el teorema de Bezout se puede represenar como un combinaci√≥n lineal del tipo**
$$
\begin{aligned}
ax+by &= d = \gcd(a,b) \\
&\text{con } a,b,x,y \in \mathbb{N}.
\end{aligned}
$$



```python

a = 400
b = 14
g,r,s = xgcd(a, b)
print(f'{g} = {r}*{a} + {s}*{b}')
# g = extended[0]
# r = extended[1]
# s = extended[2]
g == r*a + s*b
```
